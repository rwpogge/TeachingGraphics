# Taylor Series 

Numerically test a Taylor Series expansion of
 > $f(x)=c(1+x)^4$
 
and compare the full function to the first-order Taylor-series approximation
 > $f_1(x) = c(1+4x)$ 
 
derived in the *A Note About Taylor Series* handed out on Canvas.

We evaluate both functions for $c=1$ and values of $x$ ranging from 0 to 1 in steps of 0.01 and see where 
the fractional difference between the original formula and the first-order Taylor-series expansion form
 > $\frac{\delta f(x)}{f(x)} = \frac{f(x)-f_1(x)}{f(x)}$

becomes larger than 0.1%, 1%, and 10%.

```python
%matplotlib inline

import math
import numpy as np

x = np.linspace(0.0,1.0,101)

c = 1

fullForm = c*(1+x)**4
firstOrd = c*(1+4*x)
fracDiff = (fullForm-firstOrd)/fullForm

tenthPct = False
onePct = False
tenPct = False
print(' x     full   1stOr  df/f')
print('---------------------------')
for i in range(len(x)):
    if fracDiff[i] > 0.001 and not tenthPct:
        print('------- df/f > 0.1% -------')
        tenthPct = True
    if fracDiff[i] > 0.01 and not onePct:
        print('------- df/f > 1%   -------')
        onePct = True
    if fracDiff[i] > 0.1 and not tenPct:
        print('------- df/f > 10%  -------')
        tenPct = True
    print(f'{x[i]:.2f} {fullForm[i]:7.4f} {firstOrd[i]:7.4f} {fracDiff[i]:.4f}')
print('---------------------------')
```

## Graphical comparison

Plot the full and first-order Taylor series approximation functions computed above in the top panel, then
compute the fractional difference


```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots()

plt.xlim(0,1.0)
plt.plot(x,fullForm,'-',color='black',lw=1,label=r'Full Form $f(x)=(1+x)^4$')
plt.plot(x,firstOrd,'--',color='blue',lw=1,label=r'1st Order $f_1(x)\approx (1+4x)$')
plt.xlabel('x')
plt.ylabel('$f(x)$ and $f_1(x)$')
plt.legend()
plt.show()

fig,ax = plt.subplots()

plt.xlim(0,1.0)
plt.plot(x,fracDiff,'-',color='black',lw=1)
plt.xlabel('x')
plt.ylabel(r'fractional difference $\delta f(x) / f(x)$')
plt.hlines([0.01,0.1],0.0,1.0,lw=0.5,ls=[':'],colors=['black'])
plt.text(1.0,0.01,' fracDiff=1%',va='center')
plt.text(1.0,0.1,' fracDiff=10%',va='center')

plt.show()
```
